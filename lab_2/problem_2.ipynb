# Problem 2

a. Implementing $\mathrm{cos}(\theta)+j\mathrm{sin}(\theta)$ as a python function `f(x)`.

```python
import math

def f(x):
    return math.cos(math.radians(x))+math.sin((math.radians(x)))*1j
```

b. Creating a list `x` of values from -180 to 180 using the `range(start,stop,step)` function.

```python
x = range(-180,180,1)
print(list(x))
```

c. Using list comprehension to create `y`.

```python
y = [f(n) for n in x]
print(y)
```

d. Plotting: since the `y` list has real and imaginary numbers, the x-axis will be the real part, and the y-axis the imaginary part of each number in the list. This plots the beautiful unit circle!

```python
import matplotlib.pyplot as plt
import numpy as np

xx = [ele.real for ele in y] #cos
yy = [ele.imag for ele in y] #sin
ticks = np.arange(-1.25,1.50,0.25)
print(xx)
print(yy)

#plt.plot(x,yy)
#plt.plot(x,xx)
plt.plot(xx,yy)
plt.axis('equal')
plt.title('The Unit Circle drawn with Euler\'s formula')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.grid(True)
plt.xticks(ticks)
plt.yticks(ticks)
plt.xlim(-1.50,1.50)
plt.savefig('UnitCircle')
plt.show()

```
